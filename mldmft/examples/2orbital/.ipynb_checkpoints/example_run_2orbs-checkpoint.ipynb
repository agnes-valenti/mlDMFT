## I. NN-DMFT Loop

### 1) Set parameters for DMFT run using NNsolver
We start by picking values for $U$, $\mu$, $\beta$ and choosing an input hybridization function $\Delta$. In addition, we specify the tolarence for the DMFT convergence criterion, the maximum number of DMFT loops and the mixing parameter $\epsilon$, which is used for the DMFT self-consistency via 
$$ \Delta_{\rm new} = (1-\epsilon) \Delta_i + \epsilon \Delta_{i-1} $$

```python
Uval = 4.0 
t = 1.0
Jval_factor = 0.25
J = Uval * Jval_factor
mu = 3.0 / 2.0 * Uval - 5.0 / 2.0 * J

cfield = 1.5
#beta = 50.0

maxloops = 30   # number of DMFT iterations
etol = 1e-4     # DMFT SC loop convergence criteria, occupancy from G
epsilon = 0.4  # Delta mixing parameter
```

Next, we choose a starting hybridization function. This can either be done via TRIQS, and using the function ``DLR_to_NNinput'' to convert the format into the NN-input format, or by directly creating/loading a hybridization function in numpy and reshaping into the appropriate format. Concretely, the network takes as input an array of shape $A=(1, 2*n_{\tau}+3)$ with $A[0,:]=[\Re(\Delta_1), \Im(\Delta_1), ... \Re(\Delta_{n_{\tau}}), \Im(\Delta_{n_{\tau}}), U, \mu/U, \beta]$ . 

```python
useSemiCircular = True


from h5 import *
from triqs.gf import *
from triqs.gf.tools import *
from triqs.gf.meshes import MeshDLRImFreq
from mldmft.utils_2orbs import DLR_to_NNinput 
from triqs.plot.mpl_interface import oplot,plt

wmax_val = 10
eps_val = 1e-13
gf_struct = [('up', 2), ('down', 2)]
#dlr_freq_mesh = MeshDLRImFreq(beta=50, statistic='Fermion', w_max=wmax_val, eps=eps_val)
D = 1.0 
with HDFArchive(os.path.join(path,f'models/orb2/mesh_beta50.0.h5') ,'r') as h5:
    mesh_beta0 = h5['mesh_dlr_imtime']

taus = []
for tau in mesh_beta0:
    taus.append(tau.value)
print(taus)

Delta_imtime = BlockGf(mesh=mesh_beta0, gf_struct=gf_struct)
Delta_dlr = make_gf_dlr(Delta_imtime)
Delta_imfreq = make_gf_dlr_imfreq(Delta_dlr)
from triqs.gf import SemiCircular
Delta_imfreq["up"] << SemiCircular(D)
Delta_imfreq["down"] << SemiCircular(D)
#oplot(Delta["up"][0,1])

Delta_dlr = make_gf_dlr(Delta_imfreq)
NNinput = DLR_to_NNinput(Delta_dlr, mesh_beta0, Uval, mu, J, cfield) 
#from triqs.plot.mpl_interface import oplot
#G = make_gf_dlr_imtime(G)
#oplot(G)



print(NNinput.shape)
```

Next, we load the neural-network model and the stored checkpoints.

```python
device = torch.device('cpu')
#device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# load model
path_load = os.path.join(path, 'models/orb2/')
feedforwardnet = FeedforwardNet()
feedforwardnet.to(device)
feedforwardnet.load_state_dict(torch.load(os.path.expanduser(os.path.join(path_load, "save_6000.pth")), map_location=device, weights_only=True))
feedforwardnet.eval()
```

### 2) Solve DMFT loop
Now, we run a DMFT loop on the Bethe lattice, i.e. with the self-consistency $ \Delta_{i+1} = t^2 G_{i} $, where $t=1$.

```python
Goutput_prev = None 
Goutput_eps = None
converged = False
verbose = True
save_tofile = False 
conv_check = None

NN_current_dmft = np.copy(NNinput)
G_dmft = {}

batchsize = 1
ndlr_taus = (NN_current_dmft.shape[-1]-4)//6
conv_loop = maxloops

for iloop in range(maxloops):

        Goutput = feedforwardnet((torch.from_numpy(NN_current_dmft).float()).to(device)).detach().cpu().numpy()
        if Goutput_prev is None:
            Goutput_prev = np.copy(Goutput)
            Goutput_eps = np.copy(Goutput)
        else: 
            conv_check = np.linalg.norm(Goutput - Goutput_prev)
            if verbose:
               print(f"Iteration {iloop}, convergence check: {conv_check}")
            if (conv_check < 1.0e-5):
                print("Convergence reached at iteration ", iloop)
                converged = True
                conv_loop = iloop
                break
            Goutput_prev = np.copy(Goutput)
            Goutput_eps = (1-epsilon)*Goutput + epsilon*Goutput_prev
        outputvals = np.zeros((batchsize, ndlr_taus*6+4))
        outputvals[:,-4] = Uval
        outputvals[:,-3] = mu/Uval
        outputvals[:,-2] = J
        outputvals[:,-1] = cfield
        outputvals[:,:-4] = t**2 * np.reshape(Goutput_eps, (1, ndlr_taus*6))
        
        NN_current_dmft = outputvals #np.reshape(t**2 * Goutput, (1, ndlr_taus*2))
        outputs = np.reshape(Goutput, (Goutput.shape[0], ndlr_taus, 3*2))
        if conv_check is not None:
           G_dmft[iloop] = {"conv error": conv_check, "G": outputs}
        if save_tofile:
            np.save(f"dmft_NN/dmft_NN_it{iloop}.npy", outputs)

G_dmft["final"] = {"conv error": conv_check, "G": outputs}
G_dmft["n_its"] = iloop 

if not converged:
    print(f"DMFT loop has not converged, the final convergence check error is {conv_check}. \
    Change mixing parameter $\epsilon$, increase the number of maximum iterations or try a \
    different initial hybridization function $\Delta$. ")

if save_tofile:
    if not os.path.exists("dmft_NN"):
        os.makedirs("dmft_NN")
    
    np.save("dmft_NN/dmft_NN_final.npy", outputs)
    np.save("dmft_NN/beta.npy", beta)
    np.save("dmft_NN/mu.npy", mu)
    np.save("dmft_NN/U.npy", Uval)
        
```

### 3) Plot results 

```python
import matplotlib.pyplot as plt
```

```python

fig, ax = plt.subplots(2, 2, figsize=(6.5,6), dpi=100)

arange_steps = conv_loop//4
for iloop in range(1, G_dmft["n_its"]):
    if iloop in np.arange(1,conv_loop,arange_steps):
        ax[0,0].plot(taus, G_dmft[iloop]["G"][0,:,0], color = 'black', alpha = iloop/G_dmft["n_its"]*0.8 + 0.2)
        ax[0,1].plot(taus, G_dmft[iloop]["G"][0,:,1], color = 'black', alpha = iloop/G_dmft["n_its"]*0.8 + 0.2, label=f'it-{iloop}')
        ax[1,0].plot(taus, G_dmft[iloop]["G"][0,:,1], color = 'black', alpha = iloop/G_dmft["n_its"]*0.8 + 0.2)
        ax[1,1].plot(taus, G_dmft[iloop]["G"][0,:,2], color = 'black', alpha = iloop/G_dmft["n_its"]*0.8 + 0.2)

ax[0,0].plot(taus, G_dmft["final"]["G"][0,:,0], color = 'red')
ax[0,1].plot(taus, G_dmft["final"]["G"][0,:,1], color = 'red', label='final')
ax[1,0].plot(taus, G_dmft["final"]["G"][0,:,1], color = 'red')
ax[1,1].plot(taus, G_dmft["final"]["G"][0,:,2], color = 'red')

#ax.set_xlabel(r"$\tau/\beta$") 
#ax.set_ylabel(r"$G(\tau)$")
ax[0,1].legend(loc='center', bbox_to_anchor=(1.2,0.5))
for i in range(2):
    for j in range(2):
        ax[i,j].set_xlabel(r'$\tau$')
        ax[i,j].set_ylim(np.min(G_dmft["final"]["G"][0,:,:])-0.05,0.05)
        ax[i,j].text(20, -0.5, f"m={i}, m'={j}")
        if j==0:
           ax[i,j].set_ylabel(r"$G_{mm'}(\tau)$")
plt.tight_layout()
plt.savefig("Gtau.png")
```
